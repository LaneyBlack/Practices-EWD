```python
import pandas as pd

df = pd.read_csv("boston.csv")
df
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures

# Model wielomianowy (np. RM^2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(df[['rm']])
X_poly_df = pd.DataFrame(X_poly, columns=poly.get_feature_names_out(['rm']))
X_poly_df = sm.add_constant(X_poly_df)

model_poly = sm.OLS(df['medv'], X_poly_df).fit()
model_poly.summary()
```

```python
df_interact = df[['rm', 'lstat']].copy()
df_interact['rm_lstat'] = df['rm'] * df['lstat']
X_int = sm.add_constant(df_interact)

model_int = sm.OLS(df['medv'], X_int).fit()
model_int.summary()
```

```python
df_interact = df[['rm', 'lstat', 'tax']].copy()
df_interact['rm_lstat_tax'] = df['rm'] * df['lstat'] * df['tax']
X_int = sm.add_constant(df_interact)

model_int = sm.OLS(df['medv'], X_int).fit()
model_int.summary()
```

```python
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

# Select variables for interaction
features = df[['rm', 'lstat', 'ptratio']]

# Generate up to 3rd degree interactions (only interaction terms, no powers like X1^2)
poly = PolynomialFeatures(degree=3, interaction_only=True, include_bias=False)
X_interactions = poly.fit_transform(features)

# Get column names
feature_names = poly.get_feature_names_out(features.columns)

# Create DataFrame and fit model
X_df = pd.DataFrame(X_interactions, columns=feature_names)
X_df = sm.add_constant(X_df)
y = df['medv']

model = sm.OLS(y, X_df).fit()
model.summary()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm


def adjusted_r2(r2, n, k):
    """Oblicza Adjusted R^2"""
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)


def forward_selection(X, y):
    remaining = list(X.columns)
    selected = []
    best_adj_r2 = -np.inf
    n = X.shape[0]

    while remaining:
        scores = []
        for feature in remaining:
            try_features = selected + [feature]
            X_with_const = sm.add_constant(X[try_features])
            model = sm.OLS(y, X_with_const).fit()
            r2 = model.rsquared
            adj = adjusted_r2(r2, n, len(try_features))
            scores.append((adj, feature))

        scores.sort(reverse=True)
        best_new_adj_r2, best_feature = scores[0]

        if best_new_adj_r2 > best_adj_r2:
            selected.append(best_feature)
            remaining.remove(best_feature)
            best_adj_r2 = best_new_adj_r2
            print(f"Dodano: {best_feature}, Adjusted RÂ²: {best_new_adj_r2:.4f}")
        else:
            break

    return selected
```

```python
import pandas as pd
import statsmodels.api as sm

# Wczytanie danych z CSV
boston = pd.read_csv("boston.csv")

# ZakÅ‚adamy, Å¼e kolumna 'medv' to target
X = boston.drop(columns='medv')
y = boston['medv']

# Forward selection
wybrane_cechy = forward_selection(X, y)

# Budowa modelu koÅ„cowego
X_selected = sm.add_constant(X[wybrane_cechy])
model = sm.OLS(y, X_selected).fit()

# Podsumowanie modelu
print("\nPodsumowanie koÅ„cowego modelu:")
model.summary()

```

```python
import pandas as pd

df = pd.read_csv("housing.csv")

df.head()
```

```python
housing_cleaned = df.dropna()

housing_filled = df.copy()
for col in housing_filled.columns:
    if housing_filled[col].isna().sum() > 0:
        housing_filled[col] = housing_filled[col].fillna(housing_filled[col].mean())
        
housing_filled.info()
```

```python
# Zakoduj ocean_proximity
housing_encoded = pd.get_dummies(df, columns=["ocean_proximity"], drop_first=True)

# ZamieÅ„ spacje w nazwach kolumn
housing_encoded.columns = housing_encoded.columns.str.replace(" ", "_")

X.dtypes
```

```python
housing_encoded.info()
```

```python
import statsmodels.api as sm

# Select only the dummy columns for 'ocean_proximity' (and make sure to drop one to avoid multicollinearity)
X = housing_encoded[['ocean_proximity_INLAND', 'ocean_proximity_ISLAND', 'ocean_proximity_NEAR_BAY', 'ocean_proximity_NEAR_OCEAN']]

# Convert bool columns to float
X = X.astype(float)

# Target variable (median house value)
y = housing_filled['median_house_value']

# Add constant (intercept) to the independent variables
X = sm.add_constant(X)

# Build and fit the OLS model
model = sm.OLS(y, X).fit()

# Display the summary of the model
model.summary()
```
