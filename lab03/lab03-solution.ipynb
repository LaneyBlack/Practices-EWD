```python
import pandas as pd
import matplotlib.pyplot as plt

# Zadanie 1
df = pd.read_csv("data/SP500.csv", parse_dates=True, index_col="Date")
df
```

```python
plt.hist(df.Volume,bins=100, edgecolor='black')
```

```python
# Tworzenie wykresu pudełkowego
plt.figure(figsize=(6, 5))
plt.boxplot(df['Volume_M'], vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.xlabel('Day Value (millions)')
plt.title('Day Values SP500')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
df['Volume_M'] = df['Volume'] / 1e6

# Histogram
plt.figure(figsize=(10, 5))
plt.hist(df['Volume_M'], bins=30, edgecolor='black', density=True, alpha=0.7, color='blue')
plt.xlabel('Day Value (millions)')
plt.ylabel('Quantity')
plt.title('Day Values SP500')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Zadanie 2
df = pd.read_csv("data/waw_weather_monthly.csv")
df
```

```python
df['temperature_amplitude'] = df['temperature_2m_max'] - df['temperature_2m_min']

plt.bar(df["time"], df["temperature_amplitude"])
plt.xticks(rotation=45)
plt.show()
```

```python
# Zadanie 3
df = pd.read_csv("data/world_bank_co2.csv")
df
```

```python
plt.bar(df["time"], df["temperature_amplitude"], bottom=df['temperature_2m_min'])
plt.xticks(rotation=45)
plt.ylim(df['temperature_2m_min'].min()-2, df['temperature_2m_max'].max() + 2)
plt.xlabel('Month')
plt.ylabel('Temperature Range (°C)')
plt.title('Temperature Range for Each Month')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add annotations for Tmin and Tmax
for i, (month, tmin, tmax) in enumerate(zip(df['time'], df['temperature_2m_min'], df['temperature_2m_max'])):
    plt.annotate(f'{tmin:.1f}', xy=(i, tmin - 1), ha='center', fontsize=10, color='blue')
    plt.annotate(f'{tmax:.1f}', xy=(i, tmax + 1), ha='center', fontsize=10, color='red')

plt.show()
```

```python
import seaborn as sns

sns.scatterplot(
    data=df, 
    x='GDP/capita, PPP', 
    y='CO2/capita', 
    hue='Region', 
    size='Population', 
    alpha=0.4
)
plt.xlabel('GDP per Capita (PPP)')
plt.ylabel('CO2 Emissions per Capita')
plt.title('CO2 Emissions vs GDP per Capita by Region')
plt.legend(title='Region')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
df = pd.read_csv("data/iris.csv")
df
```

```python
fig, axes = plt.subplots(4, 4, figsize=(12, 12), sharex='col', sharey='row')

features = df['variety']


# Plot the scatter plots
for i in range(4):
    for j in range(4):
        if i != j:
            # Scatter plot, color by variety
            for variety in df['variety'].unique():
                subset = df[df['variety'] == variety]
                axes[i, j].scatter(subset.iloc[:, j], subset.iloc[:, i], alpha=0.5, label=variety)
        
        if i == 3:
            axes[i, j].set_xlabel(features[j])
        if j == 0:
            axes[i, j].set_ylabel(features[i])
        if i == 0:
            axes[i, j].set_title(features[j])

# Add a legend to show variety names
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=3)

plt.tight_layout()
plt.show()

# sns.pairplot(data=df, hue="variety", plot_kws={'alpha': 0.5})
```
